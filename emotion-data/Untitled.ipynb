```python
import pandas as pd
import numpy as np
    from sklearn.preprocessing import MultiLabelBinarizer


def to_array(string):
    string = string.replace(" ", "")
    number_list = string.split(',')
    number_list = [int(num) for num in number_list]     
    number_array = np.array(number_list)
    
    return number_array

def one_hot_encoding(ds):
    mlb = MultiLabelBinarizer()
    ds = ds.join(pd.DataFrame(mlb.fit_transform(ds.pop('emotion')),
                          columns=mlb.classes_,
                          index=ds.index))
    
    return ds


ds = pd.read_csv('XED/fi-annotated.tsv', sep='\t', header=None, names=['sentence', 'emotion'])
ds.emotion = ds['emotion'].apply(to_array)

ds = one_hot_encoding(ds)
```
